# Client

The client is Boto3’s low‑level interface.
It maps directly to AWS APIs — meaning every AWS API call = one Boto3 client method.

Client = Manual car (complete control, but more work).

## Real‑World Scenario

- Automating nightly backups of database dumps to S3
- Lambda functions interacting with DynamoDB
- DevOps scripts starting/stopping EC2 during off-hours

```python
import boto3
```

Boto3 client call method

```python
client = boto3.client('s3')
```

Create S3 Bucket

```python
client.create_bucket(Bucket='rohiverm-boto3-test')
```

Create new file: myfile.txt

```python
with open('myfile.txt', 'w') as file:
    file.write('This is a text file for upload')
```

Upload file to S3 Bucket

```python
client.upload_file(Filename='myfile.txt',
                Bucket = 'rohiverm-boto3-test',
                Key='test-upload-file')
```

Download file from S3 Bucket

```python
client.download_file(Bucket='rohiverm-boto3-test',
                    Key='test-upload-file',
                    Filename='myfile1.txt')
```

Delete file from S3 Bucket

```python
client.delete_object(Bucket='rohiverm-boto3-test',
                    Key='test-upload-file')
```
